```python
label_list = ['NWFcls_delta', 'NWFwrc_delta','ORFwc_delta']
color_list = ['b', 'r', 'g']
```

```python
# plot training distribution []
file_path = "train.csv"
data = np.genfromtxt(file_path, delimiter=',', skip_header = 1)
# data_selected = data[:,9:]
fig, ax = plt.subplots(1,3, figsize=(30, 6))
for idx in range(3):
    ax[idx].hist(data[:,9+idx], bins=np.arange(min(data[:,9+idx]), max(data[:,9+idx]) + 8, 8), alpha=0.5,  label=label_list[idx])
    ax[idx].axvline(data[:,9+idx].mean(), linestyle='dashed', color='k', alpha=0.7)
    min_ylim, max_ylim = ax[idx].get_ylim()
    ax[idx].text(data[:,9+idx].mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(data[:,9+idx].mean()))
    ax[idx].legend()
plt.show()
```
