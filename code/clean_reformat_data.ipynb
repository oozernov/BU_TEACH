```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


FILE = "./ecri2022.csv"

data = pd.read_csv(FILE)
```

```python
data.sort_values(by=['StuID'])
```

```python
# only consider Tx == tx
filtered_data = data[data["Tx"] == "tx"]
```

```python
# extract data with Time == 1 or Time == 2
filtered_data = filtered_data[(filtered_data["Time"] == 1) | (filtered_data["Time"] == 2)]
filtered_data
```

```python
a = filtered_data[(filtered_data["StuID"] == 1) & (filtered_data["Time"] == 1)][["NWFcls", "NWFwrc", "ORFwc"]]
a.columns = ["NWFcls_new", "NWFwrc_new", "ORFwc_new"]
# a
b = filtered_data[(filtered_data["StuID"] == 1) & (filtered_data["Time"] == 2)]
b
pd.concat([b,a], axis = 1, ignore_index=True)
```

```python
# re arrange data
fields = ['NWFcls','NWFwrc', 'ORFwc','SAwrS','SAsrS','SAtoS','RMwidRS','RMwdaRS']
StuID_set = set(filtered_data['StuID'])
new_data = pd.DataFrame()
for stuid in StuID_set:
    old = filtered_data[(filtered_data["StuID"] == stuid) & (filtered_data["Time"] == 1)][["StuID"] + fields].copy()
    new = filtered_data[(filtered_data["StuID"] == stuid) & (filtered_data["Time"] == 2)]
    for f in fields:
        old[f + '_delta'] = new[f].values[0] - old[f].values[0] # get diff in score
    new_data = pd.concat([new_data, old], ignore_index=True)
    
    
```

```python
# drop any rows with NaN
new_data.dropna(inplace=True)
new_data.reset_index(drop=True, inplace=True)
new_data
```

```python
# split data into train/test randomly by 70/30 split
total_N = new_data.shape[0]
train_N = int(0.7*total_N)
test_N = total_N - train_N

np.random.seed(2021)
row_idx = np.arange(total_N)
np.random.shuffle(row_idx)
```

```python
train_idx = row_idx[:train_N]
test_idx = row_idx[train_N:]

train_data = new_data.iloc[train_idx]
test_data = new_data.iloc[test_idx]
```

```python
# save data
train_data.to_csv('train.csv',index=False)
test_data.to_csv('test.csv',index=False)
```
